In this first exercise I will create the training dataset for my model: the dataset will be a labeled one. That means that each training point will be mapped to a label.

```python
import numpy as np
from random import randint
from sklearn.utils import shuffle
from sklearn.preprocessing import MinMaxScaler
```

```python
train_Lables = []
train_samples = []
```

creating the dataset: this dataset is about the testing of a drug.
-2100 partecipants with age range= [13-100]
-half of the partecipants were under 65 y.o. and half above
-from the sample of the people above the 65 y.o treshold around 95% experienced side effects
-from the other sample(age<65) 95% does not experienced any side effect

1st: to display if anyone has experienced side effects: 1 (yes he experienced) otherwise 0
so the logic I'll appply to create and populate the dataset is firstly to have number not percentage: 
- so the 5% of the 2100/2 individuals is 50 individuals ( I will populate this randomly with a for loop)
-the remaining 95% will be also populated with a for loop
- so each for will handle 2 branches---) 1 will populate the young people data and the other the old ones


```python
for i in range(50):
    random_younger = randint(13,64)
    train_samples.append(random_younger)
    train_Lables.append(1)
    random_older = randint(65,100)
    train_samples.append(random_older)
    train_Lables.append(0)
for i in range(1000):
    random_younger = randint(13,64)
    train_samples.append(random_younger)
    train_Lables.append(0)
    random_older = randint(65,100)
    train_samples.append(random_older)
    train_Lables.append(1)
```

let's now check if the dataset has been correctly generated

```python
for i, j in zip(train_samples, train_Lables):
    print(i,j)
```

But the fit functin does aspects a numpy format as input... and also to get rid of any input order in the data generation process I need to shuffle the data

```python
train_Lables = np.array(train_Lables)
train_samples = np.array(train_samples)
train_Lables, train_samples = shuffle(train_Lables, train_samples)
```

Neural networks and Scikit-learn tools expect data in a 2D matrix. SO now I will scale-up the data to an appropriate scale(scaled-train-sample) 

```python
scaler = MinMaxScaler(feature_range=(0,1))
scaled_train_samples = scaler.fit_transform(train_samples.reshape(-1,1))
```
