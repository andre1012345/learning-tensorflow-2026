```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Activation, Dense
from tensorflow.keras.optimizers import Adam
```

Load the data for the model

```python
train_samples = np.load('data/train_samples.npy')
train_labels = np.load('data/train_labels.npy')
```

Defining a proper model (simple data so easy model...notice I am not acting to modify the 1st layer with the first instruction but directly the second layer) of course the last layer should classify the point given so I have just 2 neurons

```python
model = Sequential([
    Dense(units=16, input_shape=(1,), activation='relu'),
    Dense(units=32, activation='relu'), # Changed sigmoid to relu (standard for hidden layers)
    Dense(units=2, activation='softmax')
])
```
